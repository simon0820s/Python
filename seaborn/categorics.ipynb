```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
print(tips.head(4))
```

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
sns.boxenplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
sns.violinplot(data=tips,y='total_bill')
plt.show()
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,split=True)
plt.show()
```

```python
#Con Catplot se puede usar cualquier grafico
#con el atributo kind
#Este se usa para graficar variables categoricas
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='violin',col='time')
plt.show()
```
