```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
print(tips.head(2))
```

```python
#corr sirve para ver la correlación entre 2 variables
#en una estructura matricial
print(tips.corr())
```

```python
#Este grafico muestra la correlación de manera grafica
#en una estructura matricial
sns.heatmap(tips.corr(),annot=True)
plt.show()
```
