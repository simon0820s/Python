```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
print(tips.head(3))
```

```python
sns.histplot(data=tips,x='tip',hue='sex',stat='percent',multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',shade=True)
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex',stat='percent')
plt.show()
```
