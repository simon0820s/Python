```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
print(tips.head(3))
```

```python
sns.histplot(data=tips,x='tip',hue='sex',stat='percent',multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',shade=True)
plt.show()
```

```python
#displot sirve para graficos de distribucin
#usando kind, se puede cambiar el grafico a cualquiera
#de los anteriores, unicamente modificando un argumento
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.show()
```
